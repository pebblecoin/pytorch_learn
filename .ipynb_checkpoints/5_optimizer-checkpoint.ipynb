```python
import torch
```

```python
import matplotlib.pyplot as plt
```

```python
import torch.utils.data as Data
```

```python
import torch.nn.functional as F
```

```python
%matplotlib inline
```

```python
x=torch.unsqueeze(torch.linspace(-1,1,1000),dim=1)
```

```python
y=x.pow(2)+0.1*torch.normal(torch.zeros(*x.size()))
```

```python
plt.scatter(x,y)
```

```python
LR = 0.01
BATCH_SIZE = 12
EPOCH = 12
```

```python
torch_dataset= Data.TensorDataset(data_tensor=x,target_tensor=y)
```

```python
loader = Data.DataLoader(dataset=torch_dataset,batch_size=BATCH_SIZE,shuffle=True,num_workers=2)
```

```python
class Net(torch.nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.hidden = torch.nn.Linear(1,20)
        self.predict = torch.nn.Linear(20,1)
    def forward(self,x):
        x=F.relu(self.hidden(x))
        x=self.predict(x)
        return x
```

```python
net_SGD = Net()
net_Momentum = Net()
net_RMSprop = Net()
net_Adam = Net()
```

```python
opt_SGD = torch.optim.SGD(net_SGD.parameters(),lr=LR)
opt_Momentum = torch.optim.SGD(net_Momentum.parameters(),lr=LR,momentum=0.7)
opt_RMSprop = torch.optim.RMSprop(net_RMSprop.parameters(),lr=LR,alpha=0.7)
opt_Adam = torch.optim.Adam(net_Adam.parameters(),lr=LR,betas=(0.9,0.99))
```

```python
nets = [net_SGD, net_Momentum, net_RMSprop, net_Adam]
```

```python
optimizers=[opt_SGD,opt_Momentum,opt_RMSprop,opt_Adam]
```

```python
loss_func = torch.nn.MSELoss()
```

```python
loss_his = [[],[],[],[]]
```

```python
from torch.autograd import Variable
```

```python
for epoch in range(EPOCH):
    print ('Epoch',epoch)
    for step,(batch_x,batch_y) in enumerate(loader):
        b_x = Variable(batch_x)
        b_y = Variable(batch_y)
        for net,opt,l_his in zip(nets,optimizers,loss_his):
            output = net(b_x)
            loss = loss_func(output,b_y)
            opt.zero_grad()
            loss.backward()
            opt.step()
            l_his.append(loss.data[0])
```

```python
labels = ['SGD', 'Momentum', 'RMSprop', 'Adam']
plt.figure(figsize=(12,10))
for i, l_his in enumerate(loss_his):
    plt.plot(l_his, label=labels[i])
plt.legend(loc='best')
plt.xlabel('Steps')
plt.ylabel('Loss')
plt.ylim((0, 0.2))
```

# Take Away:
Adam is not always the best algorithm. Momentum is better than SGD.
Also need to pay attention to the initiliazation of the network, which will also make the optimization different.
We can also try the sequential method plot.

```python
net_seq = torch.nn.Sequential(
    torch.nn.Linear(1,10),
    torch.nn.ReLU(),
    torch.nn.Linear(10,1)
)
```

```python
def init_weights(m):
    if type(m) == torch.nn.Linear:
        m.weight.data.fill_(1.0)
```

```python
net_seq.apply(init_weights)
```

```python
opt_SGD = torch.optim.SGD(net_seq.parameters(),lr=LR)
opt_Momentum = torch.optim.SGD(net_seq.parameters(),lr=LR,momentum=0.7)
opt_RMSprop = torch.optim.RMSprop(net_RMSprop.parameters(),lr=LR,alpha=0.7)
opt_Adam = torch.optim.Adam(net_seq.parameters(),lr=LR,betas=(0.9,0.99))
```

```python
l_test_his=[]
```

```python
for epoch in range(EPOCH):
    print ('Epoch',epoch)
    for step,(batch_x,batch_y) in enumerate(loader):
        b_x = Variable(batch_x)
        b_y = Variable(batch_y)
        output = net_seq(b_x)
        loss = loss_func(output,b_y)
        opt_Momentum.zero_grad()
        loss.backward()
        opt_Momentum.step()
        l_test_his.append(loss.data[0])
```

```python
 plt.plot(l_test_his)
```
